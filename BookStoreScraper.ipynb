```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup

url_template = "http://books.toscrape.com/catalogue/page-{}.html"
img_url_template = "http://books.toscrape.com"
img_urls = []
book_titles = []

with open('file.csv', 'w', encoding='utf-8') as f:
    heading = 'image_url' + ',' + 'book_title' + ',' + 'product_price' + '\n'
    f.write(heading)
    
    for i in range(1, 6):
        url = url_template.format(i)
        data = urlopen(url)
        page_html = data.read()
        page_parsed_html = soup(page_html, 'html.parser')
        books = page_parsed_html.findAll('article', {'class': 'product_pod'})

        for j in range(0, len(books)):
            book_url = books[j].img.attrs['src']
            book_title = books[j].findAll('a')[1].attrs['title']
            book_price = books[j].findAll('p')[1].text
            bookimage_url = img_url_template + book_url[2:]
            img_urls.append(bookimage_url)
            book_titles.append(book_title)
            
            if '"' in book_title:
                book_title=book_title.replace('"','')
            
            if ',' in book_url:
                book_url = '"'+book_url+'"'
            
            if ',' in book_title:
                book_title = '"'+book_title+'"'
            
            if ',' in book_price:
                book_price = '"'+book_price+'"'
            
            
            s=book_url+','+book_title+','+book_price+'\n'
            f.write(s)

data.close()
```
